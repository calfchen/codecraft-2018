```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
plt.rcParams['font.sans-serif']=['SimHei']
plt.rcParams['axes.unicode_minus']=False
```

```python
#导入相关文件
data = pd.read_table("../data/test.txt", header = None)
data.columns = ['id','kind','time']
```

```python
data.head()
```

```python
del data['id']
```

```python
data['time'] = data['time'].apply(lambda x :pd.to_datetime(x[:10], format = '%Y-%m-%d'))
data['cnt'] = 1
data = data.groupby(['kind','time']).sum().reset_index()
data['kind_num'] = data['kind'].apply(lambda x: int(x[6:]))
data = data.sort_values(by = ['kind_num','time'])
data.index = list(range(len(data)))
```

```python
data = data.groupby('kind').sum().reset_index()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
plt.rcParams['font.sans-serif']=['SimHei']
plt.rcParams['axes.unicode_minus']=False
```

```python
#导入相关文件
data = pd.read_csv("../data/data.csv")
```

```python
data.head()
```

```python
data['time'] = data['time'].apply(lambda x: pd.to_datetime(x,format="%Y-%m-%d"))
data = data.sort_values(by=['time'])
```

```python
#训练数据给3,4月份，测试数据预测5月份的7天，14天两种
train = data[data['time'] >= pd.to_datetime("2015-03-01", format="%Y-%m-%d")] [data['time'] <= pd.to_datetime("2015-04-30", format="%Y-%m-%d")]
test = data[data['time'] >= pd.to_datetime("2015-05-01", format="%Y-%m-%d")] [data['time'] <= pd.to_datetime("2015-5-31", format="%Y-%m-%d")]
```

```python
def TestDayNum(data,num):
    date = pd.to_datetime("2015-05-"+str(num+4), format = "%Y-%m-%d")
    test = data[data['time'] < date]
    test = test.groupby(['kind']).sum().reset_index()
    test = test[['kind','cnt']]
    return test
```

```python
test7 = TestDayNum(test,7)
```
